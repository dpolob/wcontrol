```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
import datetime


now = datetime.datetime.now()
now = now.replace(minute=0, hour=0, second=0, microsecond=0)

futuro = {"fecha": [],
          "temperatura": [],
          "hr": [],
          "precipitacion": []}

req = Request("https://www.eltiempo.es/najera.html?v=por_hora~ROW_NUMBER_6~~TEMP_UNIT_c~~WIND_UNIT_kmh~")
html_page = urlopen(req)

soup = BeautifulSoup(html_page, "html.parser")

html_text = soup.get_text()
```

```python
result = re.sub(r"(\n){1,10}", r"\n", html_text)
result = re.sub(r"(\t){1,10}", " ", result)
```

```python
horas = re.findall(r"\n[0-9]{1,2}:[0-9]{1,2}\n", result)
dia=0
for hora in horas:
    hora = re.sub(r"\n","", hora)
    hora = int(re.split(r":", hora)[0])
    if hora == 0:
        dia = dia + 1
    futuro['fecha'].append((now + datetime.timedelta(hours=hora, days=dia)))
futuro['fecha'].append((now + datetime.timedelta(hours=hora, days=dia)))
```

```python
temperaturas = re.findall(r"00\n[0-9]+°", result)
for temperatura in temperaturas:
    t = re.sub(r"°","", temperatura)
    t = int(re.split(r"\n", t)[1])
    futuro['temperatura'].append(int(t))
futuro['temperatura'].append(int(t))
```

```python
humedades = re.findall(r"\nHumedad\n[0-9]+%\n", result)
for humedad in humedades:
    h = re.sub(r"%", "", humedad)
    h = re.split(r"\n", h)[2]
    futuro['hr'].append(int(h))
futuro['hr'].append(int(h))
```

```python
precipitaciones = re.findall(r"\nLluvias\n[0-9]+ mm\n", result)
for precipitacion in precipitaciones:
    h = re.sub(r" mm", "", precipitacion)
    h = re.split(r"\n", h)[2]
    futuro['precipitacion'].append(int(h))
futuro['precipitacion'].append(int(h))
```
